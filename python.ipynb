### `if` statement
- `elif` and `else` parts are optional.
- When there are multiple blocks evaluated to true, the top-most block with a true condition is executed.
- values evaluated false: `None`, `False`, `0`, `0.0`, `0j`, `''`, `()`, `[]`, `{}`, `set()`, `range(0)`
- variable can be assigned a value in a condition using `:=` operator.
- if the condition is multiple lines it should be enclosed in parentheses `()`.

```python
from heapq import heapify

if (x := 2) < 0:
    print("x is less than zero.")
elif x < 10:
    print("x is less than ten.")
elif (x < 3 and
      x % 2 == 1):
    print("x is one.")
else:
    print("x is greater than equal to ten.")
```

### `while` statement
- `else` suite is executed when condition is false.
- `break` statement terminates the loop without running `else` suite.
- `continue` statement skips rest of the lines in the loop and goes back to checking condition.
- values evaluated false: `None`, `False`, `0`, `0.0`, `0j`, `''`, `()`, `[]`, `{}`, `set()`, `range(0)`
- values can be assigned to variables within a condition using `:=` operator.

```python
x = 0
while x < 5 and (y := 1):
    if y != 1:
        break
    x += 1
    if x == 3:
        continue
    print(x)
else:
    print(f"finished with x={x} and y={y}.")
```

### `for` statement
- `for` statement requires an iterable (`list`, `str`, `tuple`, `dict`, `set`, `range`) to iterate over elements.
- `break` statement terminates loop without running `else` suite.
- `continue` statement skips the rest of `for` suite and go to next iteration.

```python
for n, c in [(1, "a"), (2, "b"), (3, "c")]:
    if c == "d":
        break
    if n == 2:
        continue
    print(c)
else:
    print(f"completed for loop")
```

### `list`

#### `list` as an array
- `extend` and `+=` add each element to the end of the original list, while `append` adds the entire list to the original list.

```python
l = [1, 2]
l.append(3)
l.extend([3, 4, 5])
l += [3, 4, 5]
l.pop()
l.sort(reverse=True)
l.reverse()
len(l)
l
```

#### `list` as a stack

```python
s = [1]
s.append(2)
s.pop()
s[-1]
s
```

#### `list` as a queue

```python
from collections import deque

q = deque([1, 2])
q.append(3)
q.popleft()
q
```

### `set`
- in order to create an empty `set`, use `set()` because `{}` creates an empty dictionary.
- order is not guaranteed in a `set`.
- `set` uses `add` instead of `append`.

```python
a = set(['a', 'b', 'c'])
b = {'b', 'd'}
c = set('python')
a.add('e')
a.remove('a')
'b' in a
```

### `dict`
- keys can be either `str`, `int`, `float`, `complex`.
- list cannot be used as key.
- `del` keyword deletes a key:value pair.
- `in` keyword checks if a key exists in a `dict` returning `True` or `False`.

```python
d = {1: 'one', 2: 'two', 3: 'three'}
d[4] = 'four'
del d[2]
for key, value in d.items():
    print(key, value)
print(1 in d)
```

### `heapq`
- if initial list is not empty, heapify() should be run.
- `heapify`, `heappush`, `heappop` is min-heap by default.
- `heapify_max`, `heappush_max`, `heappop_max` is about max-heap.

```python
import heapq

minHeap = [2, 1, 3]
heapq.heapify(minHeap)
heapq.heappush(minHeap, 4)
heapq.heappop(minHeap)
minHeap[0]

maxHeap = [1, 2, 9, 5, 3]
heapq.heapify_max(maxHeap)
heapq.heappush(maxHeap, 10)
heapq.heappop(maxHeap)
maxHeap[0]
```
