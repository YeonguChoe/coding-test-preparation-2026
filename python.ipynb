### `if` statement
- `elif` and `else` parts are optional.
- When there are multiple blocks evaluated to true, the top-most block with a true condition is executed.
- variable can be assigned a value in a condition using `:=` operator.

```python
if (x := 2) < 0:
    print("x is less than zero.")
elif x < 10:
    print("x is less than ten.")
else:
    print("x is greater than equal to ten.")
```

### `while` statement
- `else` suite is executed when condition is false.
- `break` statement terminates the loop without running `else` suite.
- `continue` statement skips rest of the lines in the loop and goes back to checking condition.
- values evaluated false: `None`, `False`, `0`, `0.0`, `0j`, `''`, `()`, `[]`, `{}`, `set()`, `range(0)`
- values can be assigned to variables within a condition using `:=` operator.

```python
x = 0
while x < 5 and (y := 1):
    if y != 1:
        break
    x += 1
    if x == 3:
        continue
    print(x)
else:
    print(f"finished with x={x} and y={y}.")
```

### `for` statement
- `for` statement requires an iterable (`list`, `str`, `tuple`, `dict`, `set`, `range`) to iterate over elements.
- `break` statement terminates loop without running `else` suite.
- `continue` statement skips the rest of `for` suite and go to next iteration.

```python
for n, c in [(1, "a"), (2, "b"), (3, "c")]:
    if c == "d":
        break
    if n == 2:
        continue
    print(c)
else:
    print(f"completed for loop")
```
